```python
from graphviz import Digraph

# Initialize the Integration and Testing Diagram
integration_testing = Digraph("Integration_and_Testing", format="png", engine="dot")
integration_testing.attr(rankdir="TB", size="10")

# General Node and Edge Styling
integration_testing.attr("node", shape="box", style="filled", color="lightblue", fontcolor="black")
integration_testing.attr("edge", color="grey50")

# Main Components
integration_testing.node("UI", "User Interface", color="lightcoral")
integration_testing.node("Ontology", "Ontology (Protg)", color="gold")
integration_testing.node("Logic", "Logic Layer (Python)", color="lightgreen")
integration_testing.node("Database", "Data Storage", color="khaki")
integration_testing.node("Testing", "Testing Framework", color="lightskyblue")

# Subcomponents of Testing Framework
integration_testing.node("UnitTests", "Unit Tests", color="lightpink")
integration_testing.node("IntegrationTests", "Integration Tests", color="lightpink")
integration_testing.node("FunctionalTests", "Functional Tests", color="lightpink")

# Relationships Between Components
integration_testing.edge("UI", "Logic", label="Processes User Queries")
integration_testing.edge("Logic", "Ontology", label="Fetches Data")
integration_testing.edge("Ontology", "Database", label="Stores Metadata")
integration_testing.edge("Logic", "Database", label="Updates Records")
integration_testing.edge("UI", "Testing", label="Test Interface")
integration_testing.edge("Logic", "Testing", label="Validate Logic")
integration_testing.edge("Ontology", "Testing", label="Validate Ontology")

# Relationships Between Testing Subcomponents
integration_testing.edge("Testing", "UnitTests", label="Includes")
integration_testing.edge("Testing", "IntegrationTests", label="Includes")
integration_testing.edge("Testing", "FunctionalTests", label="Includes")

# Render and Display the Diagram
integration_testing.render(filename="Integration_and_Testing_Diagram", cleanup=True)

# Display the diagram in Google Colab
from IPython.display import Image
Image(filename="Integration_and_Testing_Diagram.png")
```
